```python
import nltk
nltk.download('vader_lexicon')
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer



```

```python
# Load the dataset
data = pd.read_csv('/content/twitter_training.csv')  # Replace with the path to your dataset

data
```

```python
# Display the first few rows of the dataset to understand its structure
print(data.head())

```

```python
# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

```

```python
# Analyze sentiment for each tweet and add the sentiment score to the DataFrame
data['Sentiment'] = data['Positive'].apply(lambda text: sia.polarity_scores(text)['compound'])

```

```python
# Plot sentiment distribution
plt.hist(data['Sentiment'], bins=30, edgecolor='black')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.title('Sentiment Distribution in Tweets')
plt.show()
```
